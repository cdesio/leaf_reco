```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
dark_folder = "/Users/cdesio/UoB/DARK-LANA/Dark/Dark"
```

```python
file0 = os.path.join(dark_folder,os.listdir(dark_folder)[0])
```

```python
img = imread(file0)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(img, cmap = "gray")
```

```python
kernel = np.ones((5,5),np.uint8)
```

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(closing, cmap='gray')
```

```python
closing2 = cv2.morphologyEx(closing, cv2.MORPH_CLOSE, np.ones((10,10),np.uint8))
```

```python
plt.figure(figsize=(10,20))
plt.imshow(closing2, cmap='gray')
```
