```python
def get_deformation(fname):
    with open(fname) as f:
        arr=[]
        dt = np.dtype([('x', np.float64), ('y', np.float64), ('z', np.float64), 
                       ('w', np.float64), ('displ', np.float64)])
        for line in f:
            if not(line.startswith("%")):
                arr.append((np.float(line.split()[0]), np.float(line.split()[1]), 
                           np.float(line.split()[2]), np.float(line.split()[6]),
                            np.float(line.split()[3])))
        deformation = np.asarray(arr, dtype=dt)
        return deformation
```

```python
deformation=get_deformation(filename)
```

```python
def plot_yz(array, xvalue, yvalue, label):
    plt.scatter(array['y'][array['x']==xvalue]-yvalue, array['w'][array['x']==xvalue],
                s=20, label=label, marker='v')
    return
```

```python
plt.figure(figsize=(12,8))
plot_yz(deformation10micron, 120.05, 140.05, '10 micron')
plot_yz(deformation100micron, 120.05, 140.05,'100 micron') 
plot_yz(deformation1mm, 120.05, 140.05,'1mm')
plot_yz(deformation2mm, 120.05, 140.05,'2mm') 
plot_yz(deformation1_75mm, 120.05,140.05, '1,75mm')
plot_yz(deformation1_75mm, 120.05, 140.05,'1,75mm')
plot_yz(deformation1_5mm, 120.05,140.05, '1,5mm')
plot_yz(deformation2_5mm, 120.05,140.05, '2,5mm')
plot_yz(deformation2_75mm, 120.05, 140.05, '2,75mm')

plt.xlabel("y position",fontsize='x-large')
plt.ylabel("deformation (mm)",fontsize='x-large')
plt.ylim(-0.5,0.01)
plt.legend()

plt.show()
```

```python
def plot_xz(array, xvalue, label):
    plt.scatter(array['x'][array['y']==xvalue]-140, array['w'][array['y']==xvalue],s=20, label=label, marker='v')
    return
```

```python
plt.figure(figsize=(12,8))
plot_xz(deformation10micron, 140, '10 micron')
plot_xz(deformation100micron, 140, '100 micron') 
plot_xz(deformation1mm, 140, '1mm')
plot_xz(deformation2mm, 140, '2mm') 
plot_xz(deformation1_75mm, 140, '1,75mm')
plot_xz(deformation1_75mm, 140, '1,75mm')
plot_xz(deformation1_5mm, 140, '1,5mm')
plot_xz(deformation2_5mm, 140, '2,5mm')
plot_xz(deformation2_75mm, 140, '2,75mm')
plt.xlabel("y position")
plt.ylabel("deformation (mm)")
plt.ylim(-0.5,0.01)
plt.legend()

plt.show()
```
