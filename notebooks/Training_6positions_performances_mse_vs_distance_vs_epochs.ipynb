```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os
```

```python
%matplotlib inline
```

```python
from sklearn.metrics import mean_squared_error
```

```python
def load_file_data(epochs):
    data = np.load("Predicted_train_6positions_{}epochs.npz".format(epochs))
    y_true = data['true']
    y_pred = data['pred']
    return(y_true, y_pred)
```

```python
def load_file_testdata(epochs):
    data = np.load("Predicted_testdata_6positions_{}epochs.npz".format(epochs))
    y_true = data['true']
    y_pred = data['pred']
    return(y_true, y_pred)
```

```python
#distances = [1,2,3,4,10,15,20,25,30,35]
distances_4pos = [1,3,15,30]
y_4_true_50, y_4_pred_50 = load_file_data(50)
y_4_true_100, y_4_pred_100 = load_file_data(100)
y_4_true_150, y_4_pred_150 = load_file_data(150)
y_4_true_200, y_4_pred_200 = load_file_data(200)
y_4_true_250, y_4_pred_250 = load_file_data(250)
y_4_true_300, y_4_pred_300 = load_file_data(300)
```

```python
distances_testdata = [2,4,10,20,25,35]
y_6_true_50, y_6_pred_50 = load_file_testdata(50)
y_6_true_100, y_6_pred_100 = load_file_testdata(100)
y_6_true_150, y_6_pred_150 = load_file_testdata(150)
y_6_true_200, y_6_pred_200 = load_file_testdata(200)
y_6_true_250, y_6_pred_250 = load_file_testdata(250)
y_6_true_300, y_6_pred_300 = load_file_testdata(300)
```

```python
def get_idx(arr, key):
    return np.where(arr==key)[0]

def idx_mse(dist, true_arr, preds_arr):
    idx = get_idx(true_arr,dist)
    mse = mean_squared_error(true_arr[idx], preds_arr[idx])
    print(dist, mse)
    return idx, mse 
```

```python
errors_50_4 = [idx_mse(i, y_4_true_50, y_4_pred_50)[1] for i in distances_4pos]
errors_100_4 = [idx_mse(i, y_4_true_100, y_4_pred_100)[1] for i in distances_4pos]
errors_150_4 = [idx_mse(i, y_4_true_150, y_4_pred_150)[1] for i in distances_4pos]
errors_200_4 = [idx_mse(i, y_4_true_200, y_4_pred_200)[1] for i in distances_4pos]
errors_250_4 = [idx_mse(i, y_4_true_250, y_4_pred_250)[1] for i in distances_4pos]
errors_300_4 = [idx_mse(i, y_4_true_300, y_4_pred_300)[1] for i in distances_4pos]
```

```python
errors_50_6 = [idx_mse(i, y_6_true_50, y_6_pred_50)[1] for i in distances_testdata]
errors_100_6 = [idx_mse(i, y_6_true_100, y_6_pred_100)[1] for i in distances_testdata]
errors_150_6 = [idx_mse(i, y_6_true_150, y_6_pred_150)[1] for i in distances_testdata]
errors_200_6 = [idx_mse(i, y_6_true_200, y_6_pred_200)[1] for i in distances_testdata]
errors_250_6 = [idx_mse(i, y_6_true_250, y_6_pred_250)[1] for i in distances_testdata]
errors_300_6 = [idx_mse(i, y_6_true_300, y_6_pred_300)[1] for i in distances_testdata]
```

```python
fig1, ax1 = plt.subplots(figsize=(16,8))
plt.title("Errors vs number of epochs - training on 6 positions",size=16)
ax1.plot(distances_4pos, errors_50_4, 's--', linewidth=1.5, c='blue',label='50 epochs_unseen')
ax1.plot(distances_4pos, errors_100_4,'s--', linewidth=1.5, c='k', label='100 epochs_unseen')
ax1.plot(distances_4pos, errors_150_4, 's--', linewidth=1.5, c='red', label='150 epochs_unseen')
ax1.plot(distances_4pos, errors_200_4, 's--', linewidth=1.5, c='green', label='200 epochs_unseen')
ax1.plot(distances_4pos, errors_250_4, 's--', linewidth=1.5, c='magenta', label='250 epochs_unseen')
ax1.plot(distances_4pos, errors_300_4, 's--', linewidth=1.5, c='orange', label='300 epochs_unseen')

ax1.plot(distances_testdata, errors_50_6, 's-', linewidth=1.5, c='blue',label='50 epochs_test')
ax1.plot(distances_testdata, errors_100_6,'s-', linewidth=1.5, c='k', label='100 epochs_test')
ax1.plot(distances_testdata, errors_150_6, 's-', linewidth=1.5, c='red', label='150 epochs_test')
ax1.plot(distances_testdata, errors_200_6, 's-', linewidth=1.5, c='green', label='200 epochs_test')
ax1.plot(distances_testdata, errors_250_6, 's-', linewidth=1.5, c='magenta', label='250 epochs_test')
ax1.plot(distances_testdata, errors_300_6, 's-', linewidth=1.5, c='orange', label='300 epochs_test')
ax1.set_ylabel("Mean Squared Error (MSE))", size=15)
ax1.set_xticks(np.arange(1,36,1))#[1,2,3,4,10,15,20,25,30,35])
ax1.set_xlabel("Leaf Positions (mm)", size=15)
ax1.set_yscale('log')
ax1.legend(loc='upper right')
plt.show()
```
