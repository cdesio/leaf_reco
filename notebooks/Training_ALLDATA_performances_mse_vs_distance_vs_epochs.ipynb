```python
from sklearn.metrics import mean_squared_error
```

```python
def load_file_data(epochs):
    data = np.load("Predicted_ALL_out_model_trained_on_ALLDATA_{}_epochs.npz".format(epochs))
    y_true = data['true']
    y_pred = data['pred']
    return(y_true, y_pred)
```

```python
distances = [1,2,3,4,10,15,20,25,30,35]
```

```python
y_true_50, y_pred_50 = load_file_data(50)
```

```python
y_true_100, y_pred_100 = load_file_data(100)
```

```python
y_true_150, y_pred_150 = load_file_data(150)
```

```python
y_true_200, y_pred_200 = load_file_data(200)
```

```python
y_true_250, y_pred_250 = load_file_data(250)
```

```python
y_true_300, y_pred_300 = load_file_data(300)
```

```python
def get_idx(arr, key):
    return np.where(arr==key)[0]

def idx_mse(dist, true_arr, preds_arr):
    idx = get_idx(true_arr,dist)
    mse = mean_squared_error(true_arr[idx], preds_arr[idx])
    print(dist, mse)
    return idx, mse 
```

```python
idx_1, mse_1 = idx_mse(1)
idx_2, mse_2 = idx_mse(2)
idx_3, mse_3 = idx_mse(3)
idx_4, mse_4 = idx_mse(4)
idx_10, mse_10 = idx_mse(10)
idx_15, mse_15 = idx_mse(15)
idx_20, mse_20 = idx_mse(20)
idx_25, mse_25 = idx_mse(25)
idx_30, mse_30 = idx_mse(30)
idx_35, mse_35 = idx_mse(35)
```

```python
errors_50 = [idx_mse(i, y_true_50, y_pred_50)[1] for i in distances]
errors_100 = [idx_mse(i, y_true_100, y_pred_100)[1] for i in distances]
errors_150 = [idx_mse(i, y_true_150, y_pred_150)[1] for i in distances]
errors_200 = [idx_mse(i, y_true_200, y_pred_200)[1] for i in distances]
errors_250 = [idx_mse(i, y_true_250, y_pred_250)[1] for i in distances]
errors_300 = [idx_mse(i, y_true_300, y_pred_300)[1] for i in distances]
```

```python
fig1, ax1 = plt.subplots(figsize=(12,8))
plt.title("Errors vs number of epochs",size=16)
ax1.plot(distances, errors_50, 'o--', linewidth=1.5, c='blue',label='50 epochs')
ax1.plot(distances, errors_100,'o--', linewidth=1.5, c='k', label='100 epochs')
ax1.plot(distances, errors_150, 'o--', linewidth=1.5, c='red', label='150 epochs')
ax1.plot(distances, errors_200, 'o--', linewidth=1.5, c='green', label='200 epochs')
ax1.plot(distances, errors_250, 'o--', linewidth=1.5, c='magenta', label='250 epochs')
ax1.plot(distances, errors_300, 'o--', linewidth=1.5, c='orange', label='300 epochs')
ax1.set_ylabel("Mean Squared Error (MSE))", size=15)
ax1.set_xlabel("Leaf Positions (mm)", size=15)
ax1.set_yscale('log')
ax1.legend(loc='upper left')
plt.show()
```
