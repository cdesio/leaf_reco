```python
def mean_std(arr1, arr2, arr3, arr4):
    mean = []
    std = []
    for a,b,c,d in zip(arr1, arr2, arr3, arr4):
        #print(a,b,c,d, np.mean([a,b,c,d]), np.std([a,b,c,d]))
        mean.append(np.mean([a,b,c,d]))
        std.append(np.std([a,b,c,d]))
    return(mean, std)
        
```

```python
mean_kapton_down_100_al, std_kapton_down_100_al = mean_std(kapton_down_100_al_1,kapton_down_100_al_2,kapton_down_100_al_3, kapton_down_100_al_4)
```
